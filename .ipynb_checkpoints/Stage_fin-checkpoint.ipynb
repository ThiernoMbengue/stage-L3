Lemmatiser et nettoyer chaques texte

```python

from nltk.stem import WordNetLemmatizer
from nltk.stem.snowball import FrenchStemmer
from collections import Counter
nltk.download('stopwords')

def is_valid_string(element):
    if element.isdigit(): #verifier si l'element est un valeur numerique
        return True
    else:
        return False


def stop_word_plus():
  L=['bien','aprés','souvent','tout','va','cala','ou','non','oui','unique','uniquement','ceux','cette','html','http','www','celle','cela','entres','faut','entre','peu','voir','comme','doit','etc','être','êtres','doit','tout','toute','tous','toutes','trop','chacun','encore','sans','autrement','fin','pour','plus','a','avoir','faire','aussi','le','la','si','ça','de','leurs','leur','autre','sous','ça','cas','dont','aucun','aucune','peut','peut','car','donc','très','droit','déja','selon','alors','où','fr', 'gyng','k']
  return L+stopwords.words('french') #les mots a supprimer



def Nettoyage_author(df):
  j=0
  texte=''
  lemmatizer = WordNetLemmatizer()
  tokenizer = nltk.RegexpTokenizer(r'\w+')
  stemmer = FrenchStemmer()
#on creer un nouveau data fram pour enregistrer les informations qu'il nous faut pour la suite
  dfFeat = pd.DataFrame(columns=['id','reponse','dictionnaire_mot','nombre_mot'])
  for i in df['author_id'].unique():
    df1=df[df['author_id']==i]
    for tx in df1['reponse']:
      texte+=' '+tx
    
    #Extraire des tokens
    texte=tokenizer.tokenize(texte)  #transformation sous forme de token
    

    #conversion en miniscule
    texte=[word.lower() for word in texte]
    #extraire les ponctuations
    texte= list(filter(lambda token: nltk.tokenize.punkt.PunktToken(token).is_non_punct, texte))
    #Remove stopwords
    texte= list(filter(lambda token: token not in stop_word_plus(), texte))

    #Lemmatisation
    #data[colonne][i]= [lemmatizer.lemmatize(word) for word in data[colonne][i]]
    
    #racinisation Stemming                                                        #A revoir si on peut faire en pyton une fonction qui fais la lemmatisation correctement
    texte= [stemmer.stem(word) for word in texte]
   
    #a revoir si on doit le mettre a la fin ou pas
    texte= list(filter(lambda token: token not in stop_word_plus(), texte))
    for mt in texte:
        if len(mt)<=2 or is_valid_string(mt)==True :
            texte.remove(mt) # on suppose tout les mots apres stem qui reste qu'un longueur inferieur a 2 est un stopword et les valeurs numerique aussi
        
     
    
    dfFeat.loc[j]=[i,texte,Counter(texte),len(texte)]
    texte=''
    j+=1
  return dfFeat

  
```

pourcentage de fois où le mot apparaît dans un texte du corpus).

```python
#df
def n_containing(word, liste_document):
  return sum(1. for document in liste_document if word in document)

```

Calculer l’indice TF-IDF d’un mot d’un texte par rapport au corpus

```python
#Realise les fonctions TF-IDF
def n_containing(word, liste_document):
  return sum(1. for document in liste_document if word in document)


def tf(word, document):   
    return document.count(word) / (len(document) * 1.)


def idf(word, liste_document):              #pour eviter de trouver O
    return math.log(len(liste_document) / (1. + n_containing(word, liste_document)))

def tfidf(word, document, liste_document):   
    return tf(word, document) * idf(word, liste_document)
```

Dans cette partie vous déterminerez un ensemble de fonctions pour :

● Déterminer l’ensemble des n mots les plus pertinents dans le corpus (n est un
paramètre, la notion de pertinence est basée sur IDF)

● Transformer un texte en un vecteur de fréquences sur les n mots les plus pertinents

● Calculer la distance cosinus entre deux textes représentés par leur vecteur de fréquence
respectif

```python

from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer

#reprendre chaque texte sous forme de string
def corpus_document(df):
    corpus=[]
    for liste in df['reponse']:
        corpus.append(" ".join(liste))
    return corpus
#pour les recuperer les n mots les plus pertinent avec leurs frequences
def get_top_n_words(corpus, n=None):
    vec = CountVectorizer().fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in     vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:n]
#recuperer que les n mots les plus pertient 
def recuper_mot_pertinent(L_tuple):
    liste_pertinent=[]
    for mt in  L_tuple:
        liste_pertinent.append(mt[0])
    return liste_pertinent
        
    
```

Transformer un texte en un vecteur de fréquences sur les n mots les plus pertinents

```python
from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer
import pandas as pd

#recuperer les texte sous form de string
def corpus_document(df):
    corpus=[]
    
    for liste in df['reponse']:
    
        
        corpus.append(" ".join(liste))
    return corpus
# on choisi les n mots les plus pertinents
def vectorisation(df,liste):
    corpus=corpus_document(df)
    # instantiate the vectorizer object
    tfidfvectorizer = TfidfVectorizer(analyzer='word',stop_words= stopwords.words('french'), vocabulary=liste)
    # convert th documents into a matrix
    tfidf_wm = tfidfvectorizer.fit_transform(corpus)

    #retrieve the terms found in the corpora
    # if we take same parameters on both Classes(CountVectorizer and TfidfVectorizer) , it will give same output of get_feature_names() methods)
    count_tokens = tfidfvectorizer.get_feature_names() # no difference
    tfidf_tokens = tfidfvectorizer.get_feature_names()
    #df_tfidfvect = pd.DataFrame(data = tfidf_wm.toarray(),index = [i for i in range(1,361)],columns = tfidf_tokens)
    df_tfidfvect = pd.DataFrame(data = tfidf_wm.toarray(),index = [i for i in range(1,len(df)+1)],columns = tfidf_tokens)
    
    return df_tfidfvect,tfidf_wm
```

```python
#méthode du coude
def coude(df,limit,title):   
    # déterminer le nombre maximum de clusters
    # en utilisant la méthode simple
    # sélection de la valeur optimale de 'k'
    # en utilisant la méthode du coude

    # wcss - dans la somme du cluster de
    # distances au carré
    wcss = {}

    for k in range(2,limit+1):
        model = KMeans(n_clusters=k)
        model.fit(df)
        wcss[k] = model.inertia_

    # tracer les valeurs wcss
    # pour connaître la valeur du coude
    plt.plot(wcss.keys(), wcss.values(), 'gs-')
    plt.xlabel('Values of "k"')
    plt.ylabel('WCSS')
    plt.title(title)
    plt.show()
```

```python
#methode silhouette
def silhouette(df,limit):
    # déterminer le nombre maximum de clusters
    # en utilisant la méthode simple
    # déterminer le nombre de clusters
    # utilisant la méthode de score de silhouette
    dic={}
    for k in range(2, limit+1):
        model = KMeans(n_clusters=k)
        model.fit(df)
        pred = model.predict(df)
        score = silhouette_score(df, pred)
        dic[str(k)]=score
    return dic
```

```python
import matplotlib
#observation
def tracer(df,k,title):
    # regrouper les données à l'aide de Kmeans
    # en utilisant k
    model = KMeans(n_clusters=k)

    # prédire les clusters
    pred = model.fit_predict(df)

    # plotting all the clusters
    colours = ['red', 'blue', 'green', 'yellow', 'orange']
    for cname in matplotlib.colors.cnames:
        colours.append(cname)
    
    # tracer tous les clusters  
    for i in np.unique(model.labels_):
        plt.scatter(df[pred==i, 0],
                    df[pred==i, 1],
                    c = colours[i])

    # tracer les centroïdes du cluster
    plt.scatter(model.cluster_centers_[:, 0], 
                model.cluster_centers_[:, 1], 
                s = 200,  # marker size
                c = 'black')

    plt.title('K Means clustering'+' '+title)

    plt.xlabel(None)
    plt.ylabel(None)
    plt.show()
```
